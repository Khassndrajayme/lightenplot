```python
"""
LightenPlot Demo Script
Demonstrates all features of the library
"""

import lightenplot as lp
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

print("=" * 60)
print("LightenPlot Demo - Simplified Data Visualization")
print("=" * 60)
```

```python
# 1. SCATTER PLOT - Single Line
print("\n1. Scatter Plot Demo")
print("-" * 60)

# Create sample data
data = pd.DataFrame({
    'age': np.random.randint(20, 60, 100),
    'salary': np.random.randint(30000, 150000, 100),
    'experience': np.random.randint(0, 20, 100)
})

# Traditional matplotlib way (10+ lines)
# import matplotlib.pyplot as plt
# fig, ax = plt.subplots(figsize=(10, 6))
# ax.scatter(data['age'], data['salary'], alpha=0.6, c='blue')
# ax.set_xlabel('Age')
# ax.set_ylabel('Salary')
# ax.set_title('Age vs Salary')
# ax.grid(True, alpha=0.3)
# plt.tight_layout()
# plt.show()

# LightenPlot way (1 line!)
lp.scatter(data, x='age', y='salary', color='steelblue') \
  .set_title('Age vs Salary') \
  .set_labels('Age (years)', 'Salary ($)') \
  .show()

print(" Scatter plot created with method chaining")
```

```python
# 2. LINE PLOT - Time Series

print("\n2. Line Plot Demo")
print("-" * 60)

time_data = pd.DataFrame({
    'month': range(1, 13),
    'sales': [120, 135, 142, 158, 170, 165, 180, 195, 210, 200, 220, 235]
})

lp.line(time_data, x='month', y='sales', color='#FF6B6B', linewidth=2.5, marker='o') \
  .set_title('Monthly Sales Performance') \
  .set_labels('Month', 'Sales (units)') \
  .apply_theme('minimal') \
  .show()

print(" Line plot created with minimal theme")
```

```python
# 3. BAR PLOT - Categorical Data

print("\n3. Bar Plot Demo")
print("-" * 60)

category_data = pd.DataFrame({
    'product': ['A', 'B', 'C', 'D', 'E'],
    'revenue': [45000, 38000, 52000, 41000, 49000]
})

lp.bar(category_data, x='product', y='revenue', color='#4ECDC4') \
  .set_title('Revenue by Product') \
  .set_labels('Product', 'Revenue ($)') \
  .save('bar_plot.png') \
  .show()

print(" Bar plot created and saved to file")
```

```python
# 4. HISTOGRAM - Distribution Analysis

print("\n4. Histogram Demo")
print("-" * 60)

ages = np.random.normal(35, 10, 1000)

lp.histogram(ages, bins=30, color='skyblue', edgecolor='navy') \
  .set_title('Age Distribution') \
  .set_labels('Age', 'Frequency') \
  .apply_theme('default') \
  .show()

print(" Histogram created with auto-binning")
```

```python
# 5. BOX PLOT - Multiple Distributions

print("\n5. Box Plot Demo")
print("-" * 60)

box_data = pd.DataFrame({
    'Group_A': np.random.normal(100, 15, 100),
    'Group_B': np.random.normal(110, 20, 100),
    'Group_C': np.random.normal(95, 12, 100)
})

lp.boxplot(box_data, columns=['Group_A', 'Group_B', 'Group_C']) \
  .set_title('Performance Comparison') \
  .set_labels('Groups', 'Score') \
  .show()

print(" Box plot comparing multiple groups")
```

```python
# 6. HEATMAP - Correlation Matrix

print("\n6. Heatmap Demo")
print("-" * 60)

# Create correlation matrix
corr_data = pd.DataFrame(
    np.random.randn(5, 5),
    columns=['Var1', 'Var2', 'Var3', 'Var4', 'Var5']
).corr()

lp.heatmap(corr_data, cmap='coolwarm', annot=True, fmt='.2f') \
  .set_title('Correlation Matrix') \
  .show()

print(" Heatmap with correlation values")
```

```python
# 7. PLOT COMPOSITION - Multiple Subplots

print("\n7. Plot Composition Demo")
print("-" * 60)

# Create individual plots
plot1 = lp.scatter(data, x='age', y='salary', color='blue')
plot1.set_title('Scatter')

plot2 = lp.line(time_data, x='month', y='sales', color='red')
plot2.set_title('Line')

plot3 = lp.bar(category_data, x='product', y='revenue', color='green')
plot3.set_title('Bar')

plot4 = lp.histogram(ages, bins=20, color='orange')
plot4.set_title('Histogram')

# Compose into 2x2 grid
composer = lp.compose(rows=2, cols=2)
composer.add_plot(plot1).add_plot(plot2).add_plot(plot3).add_plot(plot4)
composer.save('composition.png', dpi=150)
composer.show()

print(" Composed 4 plots into 2x2 grid")
```

```python

# 8. THEMES DEMO

print("\n8. Themes Demo")
print("-" * 60)

print(f"Available themes: {lp.ThemeManager.list_themes()}")

# Dark theme example
lp.scatter(data, x='age', y='experience', color='#FF6B6B') \
  .set_title('Age vs Experience') \
  .set_labels('Age', 'Years of Experience') \
  .apply_theme('dark') \
  .show()

print(" Applied dark theme")

```

```python
# 9. OOP FEATURES DEMO

print("\n9. OOP Features Demo")
print("-" * 60)

# Demonstrate dunder methods
plot_a = lp.ScatterPlot(data)
plot_b = lp.ScatterPlot(data)

print(f"Plot representation: {repr(plot_a)}")
print(f"Plot string: {str(plot_a)}")
print(f"Plots equal? {plot_a == plot_b}")

# Demonstrate composition
exporter = lp.PlotExporter(plot_a.figure)
print(f"Exporter: {repr(exporter)}")

print("\n" + "=" * 60)
print("Demo Complete! LightenPlot simplifies visualization ")
print("=" * 60)
```
