```python
# 1. SCATTER PLOT - MPG vs Weight

print("\n" + "=" * 70)
print("1. Scatter Plot: Fuel Efficiency vs Weight")
print("=" * 70)

lp.scatter(mtcars, x='wt', y='mpg', color='steelblue', size=80, alpha=0.7) \
  .set_title('Fuel Efficiency vs Car Weight') \
  .set_labels('Weight (1000 lbs)', 'Miles Per Gallon (MPG)') \
  .show()

print("✓ Created scatter plot showing negative correlation")
print("  Insight: Heavier cars tend to have lower fuel efficiency")
```

```python
# 2. SCATTER PLOT - Color by Cylinders
print("\n" + "=" * 70)
print("2. Enhanced Scatter: Color-Coded by Engine Cylinders")
print("=" * 70)

# Create color mapping for cylinders
mtcars['cyl_color'] = mtcars['cyl'].map({4: 'green', 6: 'blue', 8: 'red'})

# Corrected call to scatter with .tolist() for the color argument
lp.scatter(mtcars, x='hp', y='mpg', color=mtcars['cyl_color'].tolist(), size=100, alpha=0.6) \
  .set_title('Horsepower vs MPG (by Cylinder Count)') \
  .set_labels('Horsepower', 'Miles Per Gallon') \
  .apply_theme('minimal') \
  .show()

print("✓ Color-coded scatter plot created")
print("  Green = 4 cylinders, Blue = 6 cylinders, Red = 8 cylinders")
```

```python
# 3. BAR PLOT - Average MPG by Cylinder

print("\n" + "=" * 70)
print("3. Bar Plot: Average Fuel Efficiency by Cylinder Count")
print("=" * 70)

# Calculate average MPG by cylinder
mpg_by_cyl = mtcars.groupby('cyl')['mpg'].mean().reset_index()

lp.bar(mpg_by_cyl, x='cyl', y='mpg', color='#4ECDC4') \
  .set_title('Average MPG by Cylinder Count') \
  .set_labels('Number of Cylinders', 'Average MPG') \
  .show()

print("✓ Bar chart showing fuel efficiency decreases with more cylinders")
```

```python
# 4. HISTOGRAM - MPG Distribution

print("\n" + "=" * 70)
print("4. Histogram: Distribution of Fuel Efficiency")
print("=" * 70)

lp.histogram(mtcars, column='mpg', bins=10, color='skyblue', edgecolor='navy') \
  .set_title('Distribution of Miles Per Gallon') \
  .set_labels('Miles Per Gallon', 'Frequency') \
  .show()

print("✓ Histogram created showing MPG distribution")
print(f"  Mean MPG: {mtcars['mpg'].mean():.2f}")
print(f"  Median MPG: {mtcars['mpg'].median():.2f}")
```

```python
# 5. BOX PLOT - MPG by Transmission Type

print("\n" + "=" * 70)
print("5. Box Plot: MPG Comparison by Transmission Type")
print("=" * 70)

# Separate manual and automatic
manual = mtcars[mtcars['am'] == 1]['mpg']
automatic = mtcars[mtcars['am'] == 0]['mpg']

box_data = pd.DataFrame({
    'Manual': manual.reset_index(drop=True),
    'Automatic': automatic.reset_index(drop=True)
})

lp.boxplot(box_data, columns=['Manual', 'Automatic']) \
  .set_title('Fuel Efficiency: Manual vs Automatic Transmission') \
  .set_labels('Transmission Type', 'Miles Per Gallon') \
  .show()

print("✓ Box plot comparing transmission types")
print(f"  Manual avg: {manual.mean():.2f} MPG")
print(f"  Automatic avg: {automatic.mean():.2f} MPG")
```

```python
# 6. HEATMAP - Correlation Matrix

print("\n" + "=" * 70)
print("6. Heatmap: Correlation Between Variables")
print("=" * 70)

# Select numeric columns and calculate correlation
numeric_cols = ['mpg', 'cyl', 'disp', 'hp', 'wt']
correlation_matrix = mtcars[numeric_cols].corr()

lp.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt='.2f') \
  .set_title('Correlation Matrix: mtcars Variables') \
  .show()

print("✓ Correlation heatmap created")
print("  Strong negative correlations: mpg vs weight, mpg vs cylinders")
```

```python
# 7. MULTIPLE PLOTS - Dashboard View

print("\n" + "=" * 70)
print("7. Dashboard: Multi-Plot Composition (2x2 Grid)")
print("=" * 70)

# Create individual plots
plot1 = lp.scatter(mtcars, x='wt', y='mpg', color='blue', alpha=0.6)
plot1.set_title('Weight vs MPG')

plot2 = lp.scatter(mtcars, x='hp', y='mpg', color='red', alpha=0.6)
plot2.set_title('Horsepower vs MPG')

plot3 = lp.histogram(mtcars, column='mpg', bins=8, color='green')
plot3.set_title('MPG Distribution')

plot4 = lp.bar(mpg_by_cyl, x='cyl', y='mpg', color='orange')
plot4.set_title('Avg MPG by Cylinders')

# Compose into dashboard
composer = lp.compose(rows=2, cols=2)
composer.add_plot(plot1).add_plot(plot2).add_plot(plot3).add_plot(plot4)
composer.save('mtcars_dashboard.png', dpi=200)
composer.show()

print("✓ Created 2x2 dashboard with multiple visualizations")
print("✓ Saved as 'mtcars_dashboard.png'")
```

```python
# 8. THEME DEMONSTRATION

print("\n" + "=" * 70)
print("8. Theme Showcase: Same Plot, Different Themes")
print("=" * 70)

print("\n→ Default Theme")
lp.scatter(mtcars, x='hp', y='mpg', color='purple', size=80) \
  .set_title('Horsepower vs MPG - Default Theme') \
  .set_labels('Horsepower', 'MPG') \
  .show()

print("\n→ Dark Theme")
lp.scatter(mtcars, x='hp', y='mpg', color='cyan', size=80) \
  .set_title('Horsepower vs MPG - Dark Theme') \
  .set_labels('Horsepower', 'MPG') \
  .apply_theme('dark') \
  .show()

print("\n→ Minimal Theme")
lp.scatter(mtcars, x='hp', y='mpg', color='coral', size=80) \
  .set_title('Horsepower vs MPG - Minimal Theme') \
  .set_labels('Horsepower', 'MPG') \
  .apply_theme('minimal') \
  .show()

print("✓ Demonstrated 3 different themes with same data")
```

```python
# 9. ADVANCED: Size and Color Encoding
print("\n" + "=" * 70)
print("9. Advanced Scatter: 4D Visualization")
print("=" * 70)

# Use size for displacement, color for cylinders
lp.scatter(
    mtcars, 
    x='hp', 
    y='mpg', 
    # 📌 Change 1: Convert the color Series to a list
    color=mtcars['cyl_color'].tolist(),
    # 📌 Change 2: Convert the size Series to a list (best practice)
    size=(mtcars['disp']/2).tolist(), 
    alpha=0.6
) \
  .set_title('Multi-Dimensional Analysis: HP, MPG, Displacement, Cylinders') \
  .set_labels('Horsepower', 'Miles Per Gallon') \
  .show()

print("✓ Created 4D visualization:")
print("  X-axis: Horsepower")
print("  Y-axis: MPG")
print("  Size: Engine Displacement")
print("  Color: Number of Cylinders")
```
