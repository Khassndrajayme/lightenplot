```python
# Quick data exploration
print("Dataset Shape:", mtcars.shape)
print("\nColumn Names:")
print(mtcars.columns.tolist())
print("\nSummary Statistics:")
mtcars.describe()
```

```python
# Traditional way would be 10+ lines
# LightenPlot way: 1 line with method chaining!

lp.scatter(mtcars, x='wt', y='mpg', color='steelblue', size=80, alpha=0.7) \
  .set_title('Fuel Efficiency vs Car Weight') \
  .set_labels('Weight (1000 lbs)', 'Miles Per Gallon (MPG)') \
  .show()
```

```python
# Calculate average MPG by cylinder count
mpg_by_cyl = mtcars.groupby('cyl')['mpg'].mean().reset_index()

lp.bar(mpg_by_cyl, x='cyl', y='mpg', color='#4ECDC4') \
  .set_title('Average MPG by Cylinder Count') \
  .set_labels('Number of Cylinders', 'Average MPG') \
  .show()

print(f"\n4-cylinder avg: {mpg_by_cyl[mpg_by_cyl['cyl']==4]['mpg'].values[0]:.2f} MPG")
print(f"6-cylinder avg: {mpg_by_cyl[mpg_by_cyl['cyl']==6]['mpg'].values[0]:.2f} MPG")
print(f"8-cylinder avg: {mpg_by_cyl[mpg_by_cyl['cyl']==8]['mpg'].values[0]:.2f} MPG")
```

```python
lp.histogram(mtcars, column='mpg', bins=10, color='skyblue', edgecolor='navy') \
  .set_title('Distribution of Miles Per Gallon') \
  .set_labels('MPG', 'Frequency') \
  .show()

print(f"Mean: {mtcars['mpg'].mean():.2f} MPG")
print(f"Median: {mtcars['mpg'].median():.2f} MPG")
print(f"Std Dev: {mtcars['mpg'].std():.2f}")
```

```python
# Compare transmission types
manual = mtcars[mtcars['am'] == 1]['mpg']
automatic = mtcars[mtcars['am'] == 0]['mpg']

box_data = pd.DataFrame({
    'Manual': manual.reset_index(drop=True),
    'Automatic': automatic.reset_index(drop=True)
})

lp.boxplot(box_data, columns=['Manual', 'Automatic']) \
  .set_title('MPG: Manual vs Automatic Transmission') \
  .set_labels('Transmission Type', 'Miles Per Gallon') \
  .show()
```

```python
# Select key variables and calculate correlations
numeric_cols = ['mpg', 'cyl', 'disp', 'hp', 'wt']
correlation = mtcars[numeric_cols].corr()

lp.heatmap(correlation, cmap='coolwarm', annot=True, fmt='.2f') \
  .set_title('Correlation Matrix: Key Variables') \
  .show()
```

```python
# Create 2x2 dashboard showing different analyses

from LightenPlot import compose


mpg_by_cyl_df = mtcars.groupby('cyl')['mpg'].mean().reset_index()

plot1 = lp.scatter(mtcars, x='wt', y='mpg', color='blue', alpha=0.6)
plot1.set_title('Weight vs MPG')

plot2 = lp.scatter(mtcars, x='hp', y='mpg', color='red', alpha=0.6)
plot2.set_title('Horsepower vs MPG')

plot3 = lp.histogram(mtcars, column='mpg', bins=8, color='green')
plot3.set_title('MPG Distribution')

plot4 = lp.bar(mpg_by_cyl_df, x='cyl', y='mpg', color='orange')
plot4.set_title('Avg MPG by Cylinders')

# Compose into a 2x2 grid
composer = lp.compose(rows=2, cols=2)
composer.add_plot(plot1)
composer.add_plot(plot2)
composer.add_plot(plot3)
composer.add_plot(plot4)

composer.save("2x2.png", dpi=150)
composer.show()

```

```python
# Same plot, different themes

print("Theme Demonstration\n")

# Default theme
print("→ Default Theme")
lp.scatter(mtcars, x='hp', y='mpg', color='purple', size=80) \
  .set_title('Horsepower vs MPG - Default Theme') \
  .show()

# Dark theme
print("\n→ Dark Theme")
lp.scatter(mtcars, x='hp', y='mpg', color='cyan', size=80) \
  .set_title('Horsepower vs MPG - Dark Theme') \
  .apply_theme('dark') \
  .show()

# Minimal theme
print("\n→ Minimal Theme")
lp.scatter(mtcars, x='hp', y='mpg', color='coral', size=80) \
  .set_title('Horsepower vs MPG - Minimal Theme') \
  .apply_theme('minimal') \
  .show()
```

## Key Findings from mtcars Analysis

### Fuel Efficiency Insights:
1. **Weight Impact**: Strong negative correlation (-0.87) between weight and MPG
   - Heavier cars consume more fuel
   
2. **Cylinder Effect**: More cylinders = lower fuel efficiency
   - 4-cylinder: ~26 MPG average
   - 6-cylinder: ~20 MPG average
   - 8-cylinder: ~15 MPG average

3. **Transmission Type**: Manual transmissions show better fuel economy
   - Manual: ~24 MPG average
   - Automatic: ~17 MPG average

4. **Horsepower Trade-off**: Higher horsepower correlates with lower MPG
   - Performance comes at the cost of fuel efficiency
